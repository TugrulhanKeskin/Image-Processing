## Gradyanlar 
- Görüntü gradyanı görüntüdeki yoğunluk(pixel değeri, intensity) veya renkteki  yönlü bir değişikliktir.
- Kenar algılamada kullanılır 

```python
import cv2 
import matplotlib.pyplot as plt

# Read in the image
img = cv2.imread("sudoku.jpg", 0)
plt.figure(), plt.imshow(img, cmap= "gray"), plt.axis("off"), plt.title("Original Image")
```

```python
# X Gradient
sobelx = cv2.Sobel(img, cv2.CV_16S, dx = 1, dy =0, ksize = 5)
plt.figure(), plt.imshow(sobelx, cmap= "gray"), plt.axis("off"), plt.title("X Gradient")
```

```python
# Y Gradient
sobely = cv2.Sobel(img, ddepth = cv2.CV_16S, dx = 0, dy = 1, ksize = 5)
plt.figure(), plt.imshow(sobely, cmap= "gray"), plt.axis("off"), plt.title("Y Gradient")
```

```python
# Laplacian Gradient
laplacian = cv2.Laplacian(img, ddepth = cv2.CV_16S)
plt.figure(), plt.imshow(laplacian, cmap= "gray"), plt.axis("off"), plt.title("Laplacian Gradient")
```
