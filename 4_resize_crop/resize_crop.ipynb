```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("bulut.png") #0 renk kanalı eğer ("",0) şeklinde yazarsan siyah beyaz olur 
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
print("resim boyutu" , img.shape) #resmin ilk boyutu

plt.figure()
plt.axis("off")
plt.imshow(img)
```

```python
#resized 
imgResized = cv2.resize(img, (200,400))
print("resized img shaped ", imgResized.shape)

plt.figure()
plt.axis("off")
plt.imshow(imgResized)
```

```python
#crop
imgCropped = img[:200, :300]

plt.figure()
plt.axis("off")
plt.imshow(imgCropped)
```
