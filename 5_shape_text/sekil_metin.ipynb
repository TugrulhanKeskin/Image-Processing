```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Resim Oluştur 

```python
img = np.zeros((512,512,3), np.uint8) # siyah bir resim
print(img.shape)

plt.figure()
plt.axis("off")
plt.imshow(img)
```

### Çizgi

```python
# (resim, başlangıç noktası, bitiş noktası, renk, kalınlık
cv2.line(img, (100,100), (100,300), (0,255,0), 3) # BGR = (0,255,0)

plt.figure()
plt.axis("off")
plt.imshow(img)
```

### Dikdörtgen 

```python
# (resim, başlangıç, bitiş, renk )
cv2.rectangle(img, (0,0), (256,256), (255,0,0), cv2.FILLED)

plt.figure()
plt.axis("off")
plt.imshow(img)
```

### Çember 

```python
# (resim, merkez, yarı çap, renk)
cv2.circle(img, (300,300), 45, (0,0,255), cv2.FILLED)

plt.figure()
plt.axis("off")
plt.imshow(img)
```

### Metin

```python
# (resim, başlangıç noktası, font, kalınlığı, renk)
cv2.putText(img, "Metin Ekleme", (3,350), cv2.FONT_HERSHEY_COMPLEX, 1, (255,255,255))

plt.figure()
plt.axis("off")
plt.imshow(img)
```
