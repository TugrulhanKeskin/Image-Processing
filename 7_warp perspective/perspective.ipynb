```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ie aktar resim
img = cv2.imread("kart.png")
plt.figure()
plt.axis("off")
plt.imshow(img)
```

```python
width = 400
height = 500

pts1 = np.float32([[206,0],[0,472],[540,148],[337,615]]) #orjinal resmin ke noktalar 
pts2 = np.float32([[0,0], [0,height], [width,0], [width,height]]) #gelmesini istediimiz konumlar

matrix = cv2.getPerspectiveTransform(pts1, pts2) # noktalar dntrme
print(matrix)
```

```python
# nihai dntrlm resim
imgOutput = cv2.warpPerspective(img, matrix, (width,height)) #dntrdmz matrisi resime atama
plt.figure()
plt.axis("off")
plt.imshow(imgOutput)
```
