```python
import cv2
import matplotlib.pyplot as plt
```

### Resmi i√ße aktar 

```python
img = cv2.imread("img1.JPG")
img =cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
#RESMIN siyah beyaz olmasƒ± deƒüil renklerin skada ifade edilmesi anlamƒ±na geliyor , siyahlar 0 beyazlar 255 arasƒ±nda deƒüi≈üiyor 

plt.figure()
plt.imshow(img, cmap ="gray" ) #siyah beyaz g√∂rebilmek i√ßin color map ekledik
plt.axis("off")
plt.show()
```

### E≈üikleme
Resmi detaylardan arƒ±ndƒ±rƒ±p ana hatlarƒ±nƒ± ortaya √ßƒ±karmak i√ßin kullanƒ±lƒ±r 

```python
_, thresh_img = cv2.threshold(img, thresh = 60, maxval = 255, type = cv2.THRESH_BINARY)

plt.figure()
plt.imshow(thresh_img, cmap ="gray")
plt.axis("off")
plt.show()
```

```python
#aynƒ± kodun sonuna INV(invert) eklediƒüimizde 
_, thresh_img2 = cv2.threshold(img, thresh = 60, maxval = 255, type = cv2.THRESH_BINARY_INV)

plt.figure()
plt.imshow(thresh_img2, cmap ="gray")
plt.axis("off")
plt.show()
```

### Adaptive Threshold 
√ústte kullandƒ±klarƒ±mƒ±zda global tresholding deƒüerleri kullandƒ±k fakat kullanƒ±caƒüƒ±mƒ±z g√∂r√ºnt√ºn√ºn farklƒ± yerlerinde farklƒ± aydƒ±nlatma deƒüerleri varsa adaptive tresholding kullanƒ±lmalƒ±

Adaptive tresholdingte aynƒ± g√∂r√ºnt√ºni√º farklƒ± b√∂lgeleri i√ßin farklƒ± e≈üik deƒüerleri kullanarak , aydƒ±nlatmasƒ± farklƒ± b√∂lgelerde avantaj saƒülƒ±yor.

- Kodun a√ßƒ±klamasƒ±
    - adaptive tresholding
    #ADAPTIVE_THRESH_MEAN_C, kullanƒ±caƒüƒ±mƒ±z e≈üikleme algoritmasƒ± y√∂ntemi c sabiti ortalamadan ya da aƒüƒ±rlƒ±klƒ± ortalamadan √ßƒ±karƒ±labilecek bi deƒüer normalde pozitif ama 0 - negatif de alƒ±nabilir.
    - #cv2.THRESH_BINARY thresh holdun tipi nasƒ±l e≈üikleme yapƒ±caz , e≈üiklemeden sonra piksellerin hangi deƒüerleri alabileceƒüini belirten ifade
    - #11 blok size pksel i√ßin e≈üik deƒüeri hesaplamak i√ßin kullanƒ±lƒ±yor bi piksel topluluƒüuna g√∂re e≈üikleme yapƒ±yorduk bu topluluƒüun 11 kom≈üusuna bakƒ±yoruz diyebiliriz

```python
thresh_img3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11,8)

plt.figure()
plt.imshow(thresh_img3, cmap="gray")
plt.axis("off")
plt.show()
```
