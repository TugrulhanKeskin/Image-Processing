```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread("img1.JPG") 
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB) 
#opencv default olarak fotoları matplotlibde bgr gösteriyor orjinal rengine dönmek için rgb dönüşümü yapmalıyız
plt.figure()
plt.axis("off")
plt.imshow(img1)
```

```python
img2=cv2.imread(("img2.JPG"))
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

plt.figure()
plt.axis("off")
plt.imshow(img2)
```

```python
#karıştırma başarılı olması için boyutları eşitledik 
img1 = cv2.resize(img1, (600,600))
plt.figure()
plt.imshow(img1)

img2 = cv2.resize(img2, (600,600))
plt.figure()
plt.imshow(img2)
```

```python
#karıştırılmış resim = alpha*img1 + beta*img2 , gama 
blended = cv2.addWeighted(src1 = img1 , alpha = 0.5, src2 = img2, beta = 0.5 , gamma = 0)
plt.figure()
plt.imshow(blended)
```
