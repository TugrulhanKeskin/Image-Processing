```python
#opencv kütüphanesini içe aktaralım
import cv2

#matplotlib kütüphanesini içe aktaralım
import matplotlib.pyplot as plt
```

```python
# resmi siyah beyaz olarak içe aktaralım
img = cv2.imread('odev1.jpg', 0)

#cvtColor ile resmi renkli olarak içe aktaralım
img2 = cv2.imread('odev1.jpg')
img_renk = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
# resmi çizdirelim
plt.figure(), plt.imshow(img, cmap = "gray"), plt.axis('off'), plt.title('Odev-1 Siyah Beyaz')
plt.figure(), plt.imshow(img_renk), plt.axis('off'), plt.title('Odev-1 Renkli')
```

```python
# resmin boyutuna bakalım
print(img.shape)
```

```python
# resmi 4/5 oranında yeniden boyutlandıralım ve resmi çizdirelim
imgResized = cv2.resize(img,(int(img.shape[1]*4/5),int(img.shape[0]*4/5)))
plt.figure(), plt.imshow(imgResized, cmap = 'gray'), plt.axis('off')
```

```python
# orijinal resme bir yazı ekleyelim mesela "kopek" ve resmi çizdirelim
cv2.putText(img,"Kopek ",(375,100),cv2.FONT_HERSHEY_COMPLEX, 1 ,(0,0,0))
plt.figure(), plt.imshow(img, cmap = 'gray'), plt.axis('off')
```

```python
# original resme 50 threshold değeri üzerindekileri beyaz yap altındakileri siyah yapalım
_, thresh_img = cv2.threshold(img, thresh = 50, maxval = 255, type = cv2.THRESH_BINARY)
plt.figure(), plt.imshow(thresh_img, cmap = 'gray'), plt.axis('off')
```

```python
# orijinal resme gaussian bulanıklaştırma uygulayalım ve resmi çizdirelim
gb = cv2.GaussianBlur(img, ksize = (5,5), sigmaX = 7)
plt.figure(), plt.imshow(gb, cmap = 'gray'), plt.axis('off')
```

```python
# orijinal resme Laplacian  gradyan uygulayalım ve resmi çizdirelim
laplacian = cv2.Laplacian(img, ddepth = cv2.CV_64F)
plt.figure(), plt.imshow(laplacian, cmap = 'gray'), plt.axis('off'), plt.title('Laplacian')
```

```python
# orijinal resmin histogramını çizdirelim
color = ("b", "g", "r")
plt.figure()

for i, c in enumerate(color):
    img_hist = cv2.calcHist([img2], channels = [i], mask = None, histSize = [256], ranges = [0,256])
    plt.plot(img_hist, color = c),plt.title("Piksellerin Renk Dağılımları")
```
