# Images in scikit-image

```python
import matplotlib.pyplot as plt
# Görselleştirme Fonksiyonu
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
# scikit-image den resim okuma
from skimage import data 
rocket_image = data.rocket()
show_image(rocket_image)
```

```python
# Resmi gri tonlamaya çevirme
from skimage import color 
grayscale = color.rgb2gray(rocket_image)
show_image(grayscale, 'Grayscale')
```

```python
# Matplotlib ile resim okuma
bulut = plt.imread("bulut.jpeg")
print(f"resmin boyutu: {bulut.shape}\nveri tipi: {type(bulut)}\npiksel sayısı: {bulut.size}")
```

```python
# Resimleri numpy ile kanallarına ayırma
import numpy as np

red = bulut[:, :, 0]
green = bulut[:, :, 1]
blue = bulut[:, :, 2]

f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(20,10))
ax1.imshow(bulut)
ax1.set_title('Original'), ax1.axis('off')
ax2.imshow(red, cmap='Reds')
ax2.set_title('Red Channel'), ax2.axis('off')
ax3.imshow(green, cmap='Greens')
ax3.set_title('Green Channel'), ax3.axis('off')
ax4.imshow(blue, cmap='Blues')
ax4.set_title('Blue Channel'), ax4.axis('off')
plt.tight_layout()
plt.show()
```

```python
# Görüntüyü Yatay Çevirme
vertically_flipped = np.flipud(bulut)
show_image(vertically_flipped, 'Yatay Çevirme')
```

```python
# Görüntüyü Dikey Çevirme
horizontally_flipped = np.fliplr(bulut)
show_image(horizontally_flipped, 'Dikey Çevirme')
```
