## Rotate - Döndürme 

```python
from skimage.transform import rotate
from skimage import data
import matplotlib.pyplot as plt

cat = data.chelsea()
# saat yönünde 45 derece döndür
cat_rotated = rotate(cat, -45) # sayı negatifse saat yönünde döner
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(cat), ax1.axis('off')
ax2.imshow(cat_rotated), ax2.axis('off')
```

## Rescaling - Ölçeklendirme
- Görüntüyü belirli bir ölçeklendirme faktörüne göre yeniden boyutlandırır.

```python
from skimage.transform import rescale

# Görüntüyü 4 kat küçült ve yeniden boyutlandır 

# anti aliasing = False yaparsak görüntüyü daha hızlı küçültür
image_rescaled_faa = rescale(cat, 1/4, anti_aliasing=False, multichannel=True)

# anti aliasing = True yaparsak görüntüyü daha yavaş küçültür
image_rescaled_taa = rescale(cat, 1/4, anti_aliasing=True, multichannel=True)

f, (a1, a2, a3) = plt.subplots(1, 3, figsize=(20, 7))
a1.imshow(cat), a1.set_title('Original')
a2.imshow(image_rescaled_faa), a2.set_title('Rescaled (anti-aliasing = False)')
a3.imshow(image_rescaled_taa), a3.set_title('Rescaled (anti-aliasing = True)')
```

Aliasing : Örtüşme
- Örtüşme, görüntünün belirli bir kısmından yayılan dalgalar veya dalgalanmalar varmış gibi görünmesini sağlar.
- Nedeni pikselleşmesinin zayıf olmasıdır.

```python
# Enlarged : Resmin boyutunu 4 katına çıkarma 
image_rescaled_enlarged = rescale(cat, 4, anti_aliasing=True, multichannel=True) 
plt.imshow(image_rescaled_enlarged)
```

## Resizing - Boyutlandırma

```python
from skimage.transform import resize 
image_resized = resize(cat, (80, 80), anti_aliasing=True)
plt.imshow(image_resized)
```

```python
# Resizing proportionally 
# Resmin boyutunu 5 katına çıkarma
height = cat.shape[0]*5
width = cat.shape[1]*5
image_resized = resize(cat, (height, width), anti_aliasing=True)
plt.imshow(image_resized)
```

```python
# Görüntüyü netleştirip wallpaper olarak kullanmak için yeniden boyutlandırma ve kaydetme
bulut = plt.imread("bulut.jpg")
image_rescaled_taa = rescale(bulut, 5, anti_aliasing=True, multichannel=True)

image_resized = resize(image_rescaled_taa, (900, 1400), anti_aliasing=True)
plt.imshow(image_resized), plt.axis('off'),plt.savefig('bulut2.jpg',dpi=1400)
```
