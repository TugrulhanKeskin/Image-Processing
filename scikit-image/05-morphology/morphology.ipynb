# Morphology 
- Bir görüntüdeki nesneleri tespit etmeye çalıştığınızda, bunu şekil gibi özelliklerine göre yapabilirsiniz.
- Morfolojik filtreleme işlemleri görüntüdeki nesnelerin biçimini ve yapısını hesaba katarak kusurları giderir.
- Özellikle binary görüntülerde kullanılır.
- Basit morfolojik işlemler:
    - Erosion : Aşındırma
    - Dilation : Genişletme
    - Opening
    - Closing
    - Morphological Gradient
    - Top Hat
    - Black Hat

# Shapes in scikit-image

```python
from skimage import morphology

square = morphology.square(12)
print(square)
```

```python
rectangle = morphology.rectangle(12, 6)
print(rectangle)
```

## Erosion in scikit-image

```python
from curses.textpad import rectangle
from skimage import morphology, color
import matplotlib.pyplot as plt

# Load the image
image = plt.imread('horse.png')
image = color.rgb2gray(image)



eroded = morphology.binary_erosion(image, square)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(image, cmap='gray'), ax1.axis('off')
ax1.set_title('Original')
ax2.imshow(eroded, cmap='gray'), plt.axis('off')
ax2.set_title('Eroded')
```

## Dilation in scikit-image

```python
dilated = morphology.binary_dilation(image)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(image, cmap='gray'), ax1.axis('off') 
ax1.set_title('Original')
ax2.imshow(dilated, cmap='gray'), plt.axis('off')
ax2.set_title('Dilated')
```

## Opening and Closing in scikit-image


```python
# Opening : erosion ardından genişletme
opening = morphology.binary_opening(image, square)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(image, cmap='gray'), ax1.axis('off')
ax1.set_title('Original')
ax2.imshow(opening, cmap='gray'), plt.axis('off')
ax2.set_title('Opening')
```

```python
# Closing : genişletme ardından erosion
closing = morphology.binary_closing(image, square)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(image, cmap='gray'), ax1.axis('off')
ax1.set_title('Original')
ax2.imshow(closing, cmap='gray'), plt.axis('off')
ax2.set_title('Closing')
```
