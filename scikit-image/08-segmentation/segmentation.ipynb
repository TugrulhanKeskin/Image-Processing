# Segmentasyon (Segmentation)
- Görüntü işlemede en temel konulardan biridir
- Amaç görüntüleri basitleştirmek için bölgelere veya bölümlere ayırmaktır.
- Görüntüleri temsili daha anlamlı ve analiz edilmesi daha kolay bir şeye dönüştürürüz
- Örneğin bir tümor bilgisayarlı tomografide analiz edilmeden önce algılanmalı ve bir şekilde görüntünün geri kalanndan izole edilmelidir.
- Thresholding basit bir segmentasyon yöntemidir. 

### 2 tür Segmentasyon Yöntemi Vardır 
- Supervised (Öğretimli) : Algoritmayı yönlendirmek için bazı ön bilgiler kullanılır.
- Unsupervised (Öğretilmemiş) : Algoritma kendisi için bir yöntem bulur.

## Super Pixels 
- Super pixel, benzer renklere veya gri seviyelere sahip bir gurup bağlantılı piksellerdir.
- Görüntüler bilgisayar tarafından basit piksel ızgaraları şeklinde algılanır ve bu nedenle görüntüdeki nesnelerin sınırları keskin değildir.Super pixellerde ise bu nesnelerin sınırları daha keskin görünür.
- Görsel izleme ve görüntü sınıflandırma gibi birçok bilgisayarla görme görevinde kullanılır.
- Daha anlamlı bölgeler elde edilir
- Hesaplama verimliliğini arttırır.

## Unsupervised Segmentasyon

### Simple Linear Iterative Clustering (SLIC)
- Basit doğrusal yinelemeli kümeleme olarak çevirilebilir. 
- K-Means kümeleme adı verilen bir makine öğrenme algoritması kullanarak görüntüyü bölümlere ayırır
- Görüntünün tüm piksel değerlerini alır ve bunları önceden tanımlanmış sayıda alt bölgeye ayırmaya çalışır

```python
import matplotlib.pyplot as plt
from skimage import data

# Load the image
image = data.coffee()

plt.imshow(image), plt.axis('off'), plt.title('Original image')
```

```python
# Bu yöntem  etiketlere ayrılmış bölgeleri döndürür.
from skimage.segmentation import slic
# Parçalardan elde edilen segmentlerin olduğu bir görüntüyü döndürmek için renk modülünden label2rgb yöntemini kullanıyoruz.
from skimage.color import label2rgb

segments = slic(image)

# segmentlerin görüntüye uygulanması
# kind = avg için segmentler super piksel segmentinlerindeki ortalama renklerle doldurulur.
segmented_image = label2rgb(segments, image, kind='avg')

plt.imshow(segmented_image), plt.axis('off'), plt.title('Segmented image')
```

```python
# Daha fazla segment istiyorsak  n segments parametresini arttırabiliriz.
segments = slic(image, n_segments=300)
# kind = overlay için segmentler orjinalden farklı renklerle döndürülür.
segmented_image = label2rgb(segments, image, kind='overlay')

plt.imshow(segmented_image), plt.axis('off'), plt.title('More Segmented image')
```
